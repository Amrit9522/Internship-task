# Exploratory Data Analysis (Terrorism data)

```python
#importing the required libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
```

```python
import os
os.chdir("D:/Pandas") #changing the working directory
data= pd.read_csv("globalterrorismdb_0718dist.csv", encoding='latin1')#Loading the data
data.head(10) # viewing the first 10 rows
```

## Renaming the variables

```python
data.rename(columns={'iyear':'Year','imonth':'Month','iday':"day",'gname':'Group','country_txt':'Country','region_txt':'Region','provstate':'State','city':'City',
    'attacktype1_txt':'Attacktype','targtype1_txt':'Targettype','weaptype1_txt':'Weapon','nkill':'kill',
     'nwound':'Wound', 'natlty1_txt':'nationality'},inplace=True)
data
```

# Rearrenging the variables

```python
data=data[['Year','Month','day','Group','Country','Region','State','City','latitude','longitude','summary','location','Attacktype','Targettype','Weapon','kill','Wound','nationality','motive','target1']]
data
```

```python
data.shape
```

```python
data.describe(include='all')
```

```python
data.isnull().sum()
```

```python
data["Wound"]=data["Wound"].fillna(0)
data["kill"]=data["kill"].fillna(0)
```

## Now lets creat a new variable called total_damage_to_lives

```python
data["total_damage_to_lives"]=data['kill']+data['Wound']
data[['Wound','kill','total_damage_to_lives']].head(5)
```

## Lets have a look at history of attacks over the years in a line plot using seaborn library

```python
number_of_years=data['Year'].unique()
plt.rcParams.update({'font.size': 50})
plt.rcParams['figure.figsize']=75,30
number_of_kills =data['Year'].value_counts()
sns.lineplot(x= number_of_years, y= number_of_kills)
plt.title('History of attacks over the years',fontsize=100)
plt.xlabel('Years',fontsize=70)
plt.ylabel('Number of Attacks',fontsize=70)
plt.show()
```

from the above graph we can see the decreasing of line after 1975 and onwards is showing the number of attack is also decreasing year after year.

## Lets see the number of people died in every region

```python
temp=data.groupby(['Region'])['total_damage_to_lives'].sum().sort_values(ascending=False)
plt.rcParams['figure.figsize']=75,30
sns.barplot(x=temp.index,y=temp.values.reshape(-1))
plt.title("Affected region",fontsize=100)
plt.xlabel('Region',fontsize=70)
plt.ylabel('Number of People Died',fontsize=70)
plt.xticks(rotation=90)
plt.show()
```

## Now lets see the countries that have been attack by most number of time.

```python
attack=data.Country.value_counts()[:10]
plt.rcParams['figure.figsize']=75,30
sns.barplot(x=attack.index,y=attack.values)
plt.title('Most Attacked Country',fontsize=60)
plt.xlabel('Country',fontsize=45)
plt.ylabel('Number of Attacks',fontsize=45)
plt.xticks(rotation=90)
plt.show()

```

## Lets have a look at target type that have been suffer most by the attack.

```python
plt.rcParams.update({'font.size': 20})
hel = data['Targettype'].value_counts().drop('Unknown')[:10]  
plt.figure(figsize = (22,8))
plt.pie(hel.values, labels=hel.index)
plt.show()
```

```python
plt.rcParams.update({'font.size': 50}) 
plt.rcParams['figure.figsize']=75,30
ty = data['Group'].value_counts()[:20]
plt.figure(figsize = (50,10))
sns.barplot(x=ty.index, y=ty.values,color ='red')
plt.xlabel('City name',fontsize = 40)
plt.ylabel('Kills',fontsize = 40)
plt.title('Groups involved in attack',fontsize = 40)
plt.xticks(rotation = 90)
plt.show()
```

```python
plt.rcParams.update({'font.size': 50}) 
plt.rcParams['figure.figsize']=75,30
sns.countplot(y=data['Attacktype'], palette='inferno')
plt.xlabel('No of attacks',fontsize = 70)
plt.ylabel('Attack type',fontsize = 70)
plt.rcParams.update({'font.size': 30})
plt.title('Attack Type',fontsize = 60)
plt.show()
```

## Cities that have been attacked by most number of time

```python
plt.rcParams.update({'font.size':15})
ty = data['City'].value_counts().drop('Unknown')[:15]
plt.figure(figsize = (22,8))
plt.pie(ty.values, labels=ty.index)
plt.show()
```
